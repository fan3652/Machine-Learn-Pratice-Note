<b>感知机学习算法的原始形式求解步骤：</b><br/>
1、选取初值w0,b0<br/>
2、在训练集中任意选取点(xi,yi)<br/>
3、如果−yi(w⋅xi+b)<=0则按照(4)式更新w,b<br/>
4、重复2直到没有被误分的点<br/>
<br/>
<b>例子：用感知机学习算法的原始形式求感知机模型f(x)=sign(w·x+b)，已知，x1,x2,x3,y1,y2,y3</b>


```python
train_datas = [[3,3,1],[4,3,1],[1,1,-1]]
```

```python
def plot_points(train_datas,w,b):
    plt.figure()
    x1 = np.linspace(0, 5, 100)  
    x2 = (-b-w[0]*x1)/w[1]
    plt.plot(x1, x2, color='r', label='y1 data')
    datas_len=len(train_datas)
    for i in range(datas_len):
        if(train_datas[i][-1]==1):
            plt.scatter(train_datas[i][0],train_datas[i][1],s=50,c='green')  
        else:
            plt.scatter(train_datas[i][0],train_datas[i][1],marker='x',s=50)  
    plt.show()
```

<b>感知机学习算法的对偶形式</b>

```python
#假设学习率是1  
def percetron_function(w,b,train_datas):
    flag = False
    count = 0
    w0 = w
    b0 = b
    for i in range(len(train_datas)):
        f = train_datas[i][-1]*(np.dot(w,train_datas[i][:2])+b)
        if f <= 0:
#             print('I = %s, f = %s'%(i+1,f))
            w = list(w + np.dot(train_datas[i][-1],train_datas[i][:2]))
            b = b + train_datas[i][-1]
            flag = False
            break
        else:
            count += 1
            if count == 3:
                flag = True
            else:
                flag = False 
#     print('w = %s, b = %s'%(w,b))
#     print('***'*30)
    plot_points(train_datas,w,b)
    if flag is True:
        return w,b
    else:
        return percetron_function(w,b,train_datas)
```

```python
w,b= percetron_function([0,0],0,train_datas)
```

<b>感知机学习算法的对偶形式求解步骤：</b><br/>
1、选取初值α=0,b=0<br/>
2、在训练集中任意选取点(xi,yi)<br/>
3、如果−yi(∑α(j)y(j)x(j)·x(i)+b)<=0,a=a+η,b=b+ηy则按照(4)式更新w,b<br/>
4、重复2直到没有被误分的点<br/>
<br/>
<b>例子：用感知机学习算法的对偶形式求感知机模型f(x)=sign(w·x+b)，已知，x1,x2,x3,y1,y2,y3</b>

```python
#计算Gram矩阵
def gram_matrix(train_datas):
    result = []
    for i in range(len(train_datas)):
        value = []
        for j in range(len(train_datas)):
            value.append(np.dot(train_datas[i][:2],train_datas[j][:2]))
        result.append(value)
    return np.matrix(result)
                
```

```python
gram = gram_matrix(train_datas)
```

```python
#假设学习率是1  对偶形式
def percetron_dul(a,b,train_datas,gram):
    flag = False
    count = 0
    a0 = a
    b0 = b
    count = 0
    for i in range(len(train_datas)):
        temp = 0
        for j in range(len(train_datas)):
            temp += a[j]*train_datas[j][-1]*gram[i,j]
        temp += b
        condition = train_datas[i][-1]*temp
        if condition <= 0 :
            a[i] = a[i] + 1
            b = b + train_datas[i][-1]
            break
        else:
            count += 1
    if count == len(train_datas):
        w_list = list(map(lambda x,y,z:np.dot(x,y[:2])*z[-1],a,train_datas,train_datas))
        add = 0
        for i in w_list:
            add += i
        return a,b,add
    else:
        return percetron_dul(a,b,train_datas,gram)
```
